```python
import numpy as np
import matplotlib.pyplot as pl

data = np.genfromtxt('figureC.txt')

select= np.array([d for d in data if d[1] < 50 if d[0] < 200])
data1= select.transpose()
pl.scatter(data1[0],data1[1],s = 1, color='k', alpha=0.75, edgecolors='none');
pl.xlabel('Time in ms')
pl.ylabel('Neuron index')
pl.show();

n, bins, patches = pl.hist(data1[0], 1000, normed=0, alpha=0.75)
pl.xlabel('Time in ms')
pl.show(); 
```
